```python
df = df[df['EmployeeName'] != 'Not provided']  
df = df[df['JobTitle'] != 'Not provided']
df = df[df['BasePay'] != 'Not Provided']  
df = df[df['OvertimePay'] != 'Not Provided']
df = df[df['OtherPay'] != 'Not Provided']  
df = df[df['Benefits'] != 'Not Provided']
df.reset_index(inplace = True , drop = True)
df
```

```python
df[df.duplicated()]
```

```python
df.info()
```

```python
obj_list = ['BasePay', 'OvertimePay','OtherPay','Benefits']
for i in obj_list:
    df[i] = pd.to_numeric(df[i], errors = 'coerce')
    
df[['Id','Year']] = df[['Id','Year']].astype('int64')
    
```

```python
df.info()
```

```python
df.describe()
```

```python
df.loc[df['Benefits'].isna(),'Benefits'] = df.loc[df['Benefits'].isna(),'Benefits'].fillna(df['TotalPayBenefits']-df['TotalPay'])
df.loc[df['BasePay'].isna(),'BasePay'] = df.loc[df['BasePay'].isna(),'BasePay'].fillna(round(df['TotalPay']-df['OtherPay']-df['OvertimePay']).abs())
```

```python
df.info()
```

```python
mask_list = ['BasePay', 'OvertimePay','OtherPay','Benefits','TotalPayBenefits']
for i in mask_list:
    mask= df[i] <=100
    df = df.drop(df.index[mask])
```

```python
df.loc[df.JobTitle.str.contains("POLICE|SHERIFF|SERGEANT"), "JobCategories"] = 'POLICE'
df.loc[df.JobTitle.str.contains("NURSE|PATIENT"), "JobCategories"] = 'NURSE'
df.loc[df.JobTitle.str.contains("CLERK"), 'JobCategories'] = "CLERK"
df.loc[df.JobTitle.str.contains("HEALTH WORKER"), 'JobCategories'] = "HEALTH WORKER"
df.loc[df.JobTitle.str.contains("FIRE"), 'JobCategories'] = "FIREFIGHTER"
df.loc[df.JobTitle.str.contains("SCHOOL|LIBRA"), 'JobCategories'] = "SCHOOL EMPLOYEES"
df.loc[df.JobTitle.str.contains("ACCOUNTANT"), 'JobCategories'] = "ACCOUNTANT"
df.loc[df.JobTitle.str.contains("TRANSIT"), 'JobCategories'] = "TRANSIT OPERATOR"
df.loc[df.JobTitle.str.contains("ENGINEER"), 'JobCategories'] = "ENGINEER"
df.loc[df.JobTitle.str.contains("CUSTODIAN"), 'JobCategories'] = "CUSTODIAN"
df.loc[df.JobTitle.str.contains("RECREATION LEADER"), 'JobCategories'] = "RECREATION LEADER"
df.loc[df.JobTitle.str.contains("GUARD"), 'JobCategories'] = "GUARD"
df.loc[df.JobTitle.str.contains("ADMINISTRATIVE ANALYST"), 'JobCategories'] = "ADMINISTRATIVE ANALYST"
df.loc[df.JobTitle.str.contains("AIDE"), 'JobCategories'] = "AIDE"
df.loc[df.JobCategories.isnull(), "JobCategories"] = df.JobTitle 
```

# visualizatin 

```python
sns.pairplot(data = df)
```

```python
top_jobs = df["JobCategories"].value_counts().reset_index()
top_jobs.columns = ["JobCategories", "count"]
top_jobs
```

```python
plt.style.use("grayscale")
fig_jobs, ax_jobs = plt.subplots(figsize = (10,6))

top_jobs[:10].plot(kind = 'barh', ax = ax_jobs)

ylabels = top_jobs["JobCategories"][:10]              #ylabel
ax_jobs.set_yticklabels(ylabels, fontsize = 13);

ax_jobs.set_xlabel("count", fontstyle = "italic") #xlabel

ax_jobs.set_title("The most common jobs in San Francisco between 2012 and 2014",          #title style
          fontweight="bold", color = 'black', fontsize=14, fontname="sans", loc = "left")

ax_jobs.grid(axis ="x", drawstyle="steps", fillstyle = "left",   #grid style
          linestyle = "--")
```

```python
value_counts_total= df['Year'].value_counts().reset_index()
value_counts_total.columns = ['Year', 'Number of people doing job']
value_counts_total=value_counts_total.sort_values(by="Year")
```

```python
value_counts_total_per= df['Year'].value_counts(normalize=True).reset_index()
value_counts_total_per.columns = ['Year', 'Percentage of people']
value_counts_total_per
```

```python
fig_per_total, ax_per_total=plt.subplots(figsize=(10,6))

mypalette=["#bfcbdb", "#800000", "#146356","#a4a2a8"]
value_counts_total_per.plot(kind="pie", y="Percentage of people", ax=ax_per_total,
                           pctdistance= 0.7, shadow = True, labels=value_counts_total_per["Year"],  labeldistance=1.1,
               fontsize = 15, colors=mypalette,
               wedgeprops = {'linewidth':0.8, 'edgecolor':'black'},
               autopct='%1.1f%%', startangle = 60)

#circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

#title
ax_per_total.set_title("Percentage of people doing theirs job in SF between 2012 and 2014",          #title style
          fontstyle='italic', 
          fontweight="bold", color = 'black', fontsize=18, fontname="sans");
```

```python
fig_jobs_total, ax_jobs_total = plt.subplots(figsize = (10,6))

value_counts_total.plot(kind = 'bar', ax = ax_jobs_total, color="#800000", legend = False)

xlabel=value_counts_total["Year"]
ax_jobs_total.set_xticklabels(xlabel)
ax_jobs_total.set_ylabel("count", fontstyle = "italic") #ylabel

ax_jobs_total.set_title("The count of people, doing jobs in San Francisco between 2012 and 2014",          #title style
          fontweight="bold", color = 'black', fontsize=14, fontname="sans", loc = "left")

ax_jobs_total.grid(axis ="x", drawstyle="steps", fillstyle = "left",   #grid style
          linestyle = "--")
```
